## Quick Reference
```sql
WITH RECURSIVE cte_name AS (
    -- Anchor: Base case (non-recursive)
    SELECT ... WHERE parent IS NULL
    
    UNION ALL
    
    -- Recursive: References itself
    SELECT ...
    FROM table t
    JOIN cte_name c ON t.parent_id = c.id
)
SELECT * FROM cte_name;
```
