## Quick Reference
```sql
-- Simple CASE
CASE column
    WHEN value1 THEN result1
    WHEN value2 THEN result2
    ELSE default_result
END

-- Searched CASE
CASE
    WHEN condition1 THEN result1
    WHEN condition2 THEN result2
    ELSE default_result
END

-- NULL handling
COALESCE(col, default_value)  -- Returns first non-NULL
NULLIF(col, value)  -- Returns NULL if col = value
```
